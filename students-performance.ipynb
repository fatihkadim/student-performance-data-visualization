```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("StudentsPerformance.csv")
df = data.copy()
```

```python
df.head()
```

##### gender: Öğrencinin cinsiyeti (male: erkek, female: kadın).
##### race/ethnicity: Öğrencinin ırkı veya etnik kökeni (ör. group A, group B).
##### parental level of education: Ebeveynlerin eğitim seviyesi (ör. high school, bachelor's degree).
##### lunch: Öğle yemeği türü (standard: standart, free/reduced: ücretsiz/indirimli).
##### test preparation course: Sınav hazırlık kursuna katılım (completed: tamamlanmış, none: katılmamış).
##### math score: Matematik sınavından alınan puan (sayısal değer).
##### reading score: Okuma sınavından alınan puan (sayısal değer).
##### writing score: Yazma sınavından alınan puan (sayısal değer).

```python
df.describe().T
```

```python
sns.barplot(x = "gender",y ="math score",data = df);
```

Erkeklerin matematik ortalaması daha yüksek.

```python
sns.barplot(x = "gender",y ="reading score",data = df);
```

Kızların okuma sınavındaki ortalamaları daha yüksek.

```python
sns.barplot(x = "gender",y ="writing score",data = df);
```

Kızların yazma sınavındaki skorları daha yüksek.

```python
sns.barplot(x = "gender",y ="math score",hue = "parental level of education",data = df);
```

**Erkekler genel olarak kızlardan daha yüksek matematik puanlarına sahip. Bu fark, özellikle ebeveynleri daha yüksek eğitimli olan öğrencilerde daha belirgin.
Ebeveynlerin eğitim seviyesi arttıkça, hem kız hem de erkek öğrencilerin matematik puanları da artıyor. Yani, ebeveynleri üniversite mezunu olan öğrencilerin puanları, ebeveynleri lise mezunu olan öğrencilere göre daha yüksek.
En büyük puan farkı, ebeveynleri üniversite veya daha yüksek eğitimli olan erkek ve kız öğrenciler arasında görülüyor. Bu, yüksek eğitimli ebeveynlerin erkek çocuklarının matematik başarısını kız çocuklarına göre daha fazla desteklediği anlamına gelebilir.**

```python
sns.barplot(x = "gender",y ="math score",hue = "test preparation course",data = df);
```

**Erkekler kızlardan daha yüksek ortalama matematik puanlarına sahip.
Sınava hazırlık kursu alan öğrenciler, almayanlara göre daha yüksek puanlar almış.
Kursun etkisi erkeklerde daha belirgin. Erkekler arasında kurs alanlar ile almayanlar arasındaki puan farkı, kızlara göre daha fazla.
Özetle, hem cinsiyet hem de kurs alma durumu matematik başarısını etkiliyor. Kurs, özellikle erkek öğrenciler için faydalı görünüyor.**

```python
sns.barplot(x = "gender",y ="math score",hue = "race/ethnicity",data = df);
```

**Farklı etnik gruplar arasında matematik puanları farklılık gösteriyor. Örneğin, "group E" hem kızlarda hem de erkeklerde en yüksek ortalama puanlara sahipken, "group A" en düşük puanlara sahip.
Cinsiyet farkı etnik gruplara göre değişiyor. Bazı gruplarda erkekler ve kızlar arasında daha büyük fark varken, bazılarında fark daha az.
Özetle, hem cinsiyet hem de etnik köken matematik başarısını etkiliyor.**

```python
sns.boxplot(x = "gender", y = "math score",data = df);
```

```python
sns.boxplot(x = "gender", y = "reading score",hue = "race/ethnicity",data = df);
```

**Üstte bu grafiği matematik sonuçlarına göre barplot yöntemiyle yorumlamıştık.Bu grafikte ise boxplot kullanarak okuma puanlarını inceledik.
Kızların ortalama okuma puanları genel olarak erkeklerden biraz daha yüksek.
Etnik kökene göre okuma puanları değişiyor; "group E" genellikle daha yüksek puanlara sahip.
Kutuların uzunluğu puanların dağılımını, çemberler ise aykırı değerleri gösteriyor.
Kısacası, hem cinsiyet hem de etnik köken okuma becerilerini etkiliyor.**

```python
sns.boxplot(x = "parental level of education",y = "math score",hue = "test preparation course",data = df);
plt.xticks(rotation=45)  
plt.tight_layout()  
```

**Ebeveynlerin eğitim seviyesi arttıkça matematik puanları genellikle artıyor.
Kurs alanların puanları, almayanlardan genellikle daha yüksek.
Kursun etkisi, ebeveynleri daha az eğitimli olan öğrencilerde daha belirgin.
Yani hem ebeveyn eğitimi hem de kurs, matematik başarısını olumlu etkiliyor.**

```python
sns.boxplot(x = "gender", y = "writing score",hue = "race/ethnicity",data = df);
```

Kadınların yazma puanlarının genelde erkeklerden daha yüksek olduğu görülüyor. "Group E" üyeleri en yüksek, "Group A" üyeleri ise en düşük puanlara sahip. Aykırı değerler özellikle düşük puanlarda dikkat çekiyor. Yazma puanları, hem cinsiyet hem de etnik gruplar arasında belirgin farklılıklar gösteriyor.

```python
sns.scatterplot(x = "reading score",y = "math score",data =df);
```

Noktalardan anlaşıldığı üzere, iki değişken arasında güçlü ve pozitif bir ilişki var; yani, okuma puanı yüksek olan öğrencilerin genelde matematik puanı da yüksek. Veriler genel olarak çapraz bir çizgi boyunca yoğunlaşmış, ancak düşük okuma puanlarında daha fazla yayılım ve bazı düşük matematik puanları dikkat çekiyor. Bu durum, okuma ve matematik başarısının birbirini desteklediğini gösteriyor.

```python
sns.scatterplot(x = "reading score",y = "math score",hue = "race/ethnicity",data =df);
```

```python
sns.scatterplot(x = "reading score",y = "math score",hue = "test preparation course",data =df);
```

Genel olarak, okuma ve matematik puanları arasında güçlü bir pozitif ilişki olduğu görülüyor. Test hazırlık kursunu tamamlayan öğrenciler (turuncu noktalar), genelde her iki puanda da daha yüksek sonuçlar almış. Test hazırlık kursuna katılmayan öğrenciler (mavi noktalar) arasında ise daha düşük puanlar gözlemleniyor. Bu, test hazırlık kursunun akademik başarı üzerinde olumlu bir etkisi olabileceğini gösteriyor.

```python
sns.scatterplot(x = "reading score",y = "math score",hue = "test preparation course",style = "gender",data =df);
```

Bu grafikte ise ekstra olarak cinsiyetler arasında farklı dağılımlar dikkat çekmektedir. Kadın öğrenciler (daire semboller) daha dengeli bir performans sergilerken, erkek öğrenciler (çarpı semboller) arasında hem çok düşük hem de çok yüksek uç değerler görülmektedir. Bu da erkek öğrencilerde performans farklılıklarının daha belirgin olduğunu, kadın öğrencilerin ise daha tutarlı bir başarı sergilediğini göstermektedir.

```python
sns.lmplot(x = "reading score",y= "math score",hue = "parental level of education",col = "race/ethnicity",row = "gender",data = df);
```

Genel Bulgular:
Okuma ve Matematik Puanı İlişkisi: Tüm grafiklerde pozitif bir eğilim görülüyor, yani okuma puanı arttıkça matematik puanı da artıyor. Bu ilişki, cinsiyet ve etnik grup fark etmeksizin tutarlı görünüyor.
Ebeveyn Eğitiminin Etkisi: Ebeveynlerin eğitim seviyesi, matematik ve okuma puanları üzerinde belirgin bir etkiye sahip. Genelde, "master's degree" (yeşil) ve "bachelor's degree" (mavi) seviyelerine sahip ebeveynlerin çocukları daha yüksek puanlara ulaşma eğiliminde.
Cinsiyet Farklılıkları: Kadın ve erkek öğrenciler arasında çok büyük bir fark olmamakla birlikte, bazı etnik gruplarda performans eğilimleri değişiyor. Örneğin, kadın öğrenciler grup D ve E'de daha dengeli ve yüksek performans gösterirken, erkek öğrenciler grup A ve C'de daha geniş bir dağılım sergiliyor.
Etnik Grupların Farklılıkları: Gruplar arasında puan eğilimlerinde bazı farklılıklar var. Örneğin, grup E öğrencileri genelde daha yüksek puanlar alırken, grup A öğrencilerinde daha geniş bir varyasyon ve düşük puanlar dikkat çekiyor.
Belirsizlik Alanları: Şeffaf gölgeler, regresyon çizgisinin etrafındaki tahmin belirsizliğini göstermektedir. Grup A ve B gibi bazı gruplarda belirsizlik daha fazla, bu da bu gruplardaki verilerin daha geniş bir dağılım sergilediğini gösteriyor.
Sonuç olarak, grafikler okuma ve matematik performansı üzerinde ebeveyn eğitimi, cinsiyet ve etnik grubun etkilerini görselleştiriyor. En yüksek performans genellikle daha yüksek ebeveyn eğitimi ve grup E'de görülürken, farklı gruplar arasında varyasyonlar dikkat çekiyor.
